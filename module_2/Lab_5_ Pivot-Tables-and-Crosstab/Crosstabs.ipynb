### 1. Import the moovies dataset

```python
movie = pd.read_csv(r'./movie_metadata.csv')
movie.head()
```

### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(movie.country, movie.color, margins=True).All
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([movie['director_name'], movie['genres']], movie['country'])
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(movie['color'], movie['color'], values=movie['duration'], aggfunc="mean", margins=True).All
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(movie['director_name'], movie['color'], values=movie['cast_total_facebook_likes'], aggfunc="mean")
```
