```python
import pandas as pd
import seaborn as sbs
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'./vehicles.csv')
data.head()
```

```python
data.boxplot(column= "Combined MPG")
```

```python
sbs.boxplot(data=data,x= "Combined MPG")
```

```python
sbs.boxplot(data=data,x= "Combined MPG", y="Fuel Type")
```

```python
sbs.boxplot(data=data,y= "Combined MPG", x="Fuel Type")
plt.xticks(rotation=90);
```

```python
cyl = [6.0 , 4.0]
data.Cylinders.isin(cyl)
data1 = data[data.Cylinders.isin(cyl)]
data1.shape
```

```python
ax = sbs.boxplot(data=data1, x= "Combined MPG", y="Fuel Type", hue="Cylinders", palette="Set3")
ax.set_xticklabels(ax.get_xticklabels(),rotation=90);
```

```python
sbs.violinplot( y="Combined MPG", x="Cylinders", data=data1)
```

```python
sbs.histplot(data=data, x="Combined MPG", bins= 5)
```

```python
sbs.histplot(data=data, x="Combined MPG", bins= 50)
```

```python
sbs.distplot(data["Combined MPG"]);
```

```python
data['Drivetrain'].value_counts().plot.bar()
```

```python
bx = sbs.barplot(data=data1, x= "Fuel Type", y="Highway MPG", hue="Cylinders", palette="Set3")
bx.set_xticklabels(bx.get_xticklabels(),rotation=90);
```
