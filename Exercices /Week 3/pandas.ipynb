```python
#vehicles
vehic=pd.read_csv('/Users/jaimesastrecrespo/DAFT_1022/Exercices /Week 3/vehicles/vehicles.csv')
vehic
```

```python
vehic.columns
```

```python
vehic = vehic.rename(columns = {'Make' : 'Producer', 'Engine Displacement' : 'Displacement'})
```

```python
vehic.columns
```

```python
column_order = ['Year','Vehicle Class','Transmission','Drivetrain','Fuel Type',
                'Cylinders','Displacement','Fuel Barrels/Year',
                'City MPG','Highway MPG','Combined MPG',
                'CO2 Emission Grams/Mile','Fuel Cost/Year']


vehic = vehic[column_order]
```

```python
vehic
```

```python
filtered = vehic[(vehic['Fuel Type']=='Regular') & 
                (vehic['Cylinders']>=6) & 
                (vehic['Fuel Cost/Year'] < 2000)]


filtered.head()
```

```python
avg_mpg = vehic.groupby('Fuel Type', as_index=False).agg({'Combined MPG':'mean'})
avg_mpg.columns = ['Fuel Type', 'Avg_MPG']
```

```python
avg_mpg
```

```python
vehic = pd.merge(vehic, avg_mpg, on='Fuel Type')
vehic.head(10)
```
