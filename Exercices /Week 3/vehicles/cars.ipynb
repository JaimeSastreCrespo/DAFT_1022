```python
# import data set from Ironhack's database
cars=pd.read_csv(r'/Users/jaimesastrecrespo/DAFT_1022/Exercices /Week 3/vehicles/vehicles.csv')
```

```python
cars.loc[cars['Transmission'].str.startswith("A"), 'Transtype'] = 'Automatic'
cars.loc[cars['Transmission'].str.startswith("M"), 'Transtype'] = 'Manual'
```

```python
cars
```

```python
avg_mpg = cars.groupby('Make').agg({'Combined MPG': 'mean'})
avg_mpg
```

```python
avg_cyl = cars.groupby('Year', as_index = False).agg({'Cylinders': 'mean'})
avg_cyl
```

```python
avg_mpg.columns = ['Make', 'Avg_MPG']
```

```python
data = pd.merge(cars, avg_mpg, on='Make')
data.head(10)
```

```python
filtered = data[(data['Make']=='Ford') & 
                (data['Cylinders']>=6) & 
                (data['Combined MPG'] < 18)]


filtered.head()
```
