```python
df1 = df.copy()
```

```python

# My code for the task. First I identified the duplicates in the name column and filled what I could.
accounts = list(df1['Account'].unique())
for orga in accounts :
    if df1.loc[df1['Account'] == orga].shape[0] > 1 :
        masterID = int(df1.loc[df1['Account'] == orga, 'AccountID'].mode().min())
        df1.loc[df1['Account'] == orga, 'MasterID'] = masterID
        domain = df1.loc[df1['Account'] == orga, 'Email domain'].mode().min()
        df1.loc[(df1['Account'] == orga) & (df1['Email domain'].isna()), 'Email domain'] = domain
        acronym = df1.loc[df1['Account'] == orga, 'Acronym'].mode().min()
        df1.loc[(df1['Account'] == orga) & (df1['Acronym'].isna()), 'Acronym'] = acronym
```

```python
# Then I tracked the duplicates by email domain with the same technique.
domains = list(df1['Email domain'].unique())
for domain in domains :
    if df1.loc[df1['Email domain'] == domain].shape[0] > 1 :
        masterID = int(df1.loc[df1['Email domain'] == domain, 'AccountID'].mode().min())
        df1.loc[df1['Email domain'] == domain, 'MasterID'] = masterID
        acronym = df1.loc[df1['Email domain'] == domain, 'Acronym'].mode().min()
        df1.loc[(df1['Email domain'] == domain) & (df1['Acronym'].isna()), 'Acronym'] = acronym
```

```python
# I did the same with acronyms, but this part yields issues since different organizations 
# share the exact same acronym (check BELGIAN PETROLEUM FEDERATION and BRITISH PROPERTY FEDERATION).
acronyms = list(df1['Acronym'].unique())
for acronym in acronyms :
    if df1.loc[df1['Acronym'] == acronym].shape[0] > 1 :
        masterID = int(df1.loc[df1['Acronym'] == acronym, 'AccountID'].mode().min())
        df1.loc[df1['Acronym'] == acronym, 'MasterID'] = masterID
```
