```python
import pandas as pd
from scipy.stats import ttest_rel
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
```

```python
blood = pd.read_csv(r'./blood_pressure.csv')
```

```python
ttest_rel(blood.after, blood.before)
```

```python
print("{:.50f}".format(ttest_rel(blood.after, blood.before).pvalue))
```

```python
from scipy.stats import ttest_1samp
```

```python
ttest_1samp(blood.after-blood.before,0)
```

```python
print("{:.50f}".format(ttest_1samp(blood.after-blood.before, 0).pvalue))
```

```python
city_rate = pd.read_csv(r'./rate_by_city.csv')
city_rate.head()
```

```python
city_rate['city_count'] = city_rate.groupby('City').cumcount()
city_rate
```

```python
rate_pivot = city_rate.pivot(index = 'city_count', columns = 'City', values = 'Rate')
rate_pivot
```

```python
rate_pivot.columns = ['City_'+str(x) for x in rate_pivot.columns.values]
```

```python
rate_pivot.head()
```

```python
f_oneway(rate_pivot.City_1,rate_pivot.City_2,rate_pivot.City_3,rate_pivot.City_4,rate_pivot.City_5,rate_pivot.City_6)
```
