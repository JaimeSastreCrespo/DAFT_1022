```python
#Drop the column Department because it's always the same department so, we assume we know where are we
df.drop(['Department'], axis=1, inplace = True)
df.drop([1], inplace=True)
```

```python
#replacing the nan values by the most frequent value of eahc column
df = df.fillna(df.mode().iloc[0])
```

```python
#Rounded the numbers by .astype as an integer
df['BirthYear'] = df['BirthYear'].astype(int)
df['ClientID'] = df['ClientID'].astype(int)
df['Amount'] = df['Amount'].astype(int)
df['TransactionID'] = df['TransactionID'].astype(int)
```

```python
df['Profession'] = df['Profession'].str.lower()
```

```python
df['Profession'].value_counts()
```

```python
risk_dict = {
    'Low' : 1,
    'Medium' : 2,
    'High' : 3
}
```

```python
df['Risk_Order'] = df['Risk'].map(risk_dict)
df.drop(columns=['Risk'], inplace=True)
df
```

```python
df.loc[df['Profession'] == 'etudient', 'Profession'] = 'student'
```
