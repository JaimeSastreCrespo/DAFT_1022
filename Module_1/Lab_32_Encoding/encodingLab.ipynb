### Label Encoder

We use label encoder to encode target labels with value between 0 and n-1.

```python
lab_enc = LabelEncoder()
```

```python
lab_df = df
lab_df['Profession'] = lab_enc.fit_transform(df['Profession'])
lab_df['Risk'] = lab_enc.fit_transform(df['Risk'])
lab_df['Bank_dep'] = lab_enc.fit_transform(df['Bank_dep'])
lab_df
```
