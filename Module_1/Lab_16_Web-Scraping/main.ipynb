#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')

res = soup.find_all("h1", class_="h3 lh-condensed")


for i in res:
    print(i.contents[1]['href'])

````

#### Display all the image links from Walt Disney wikipedia page.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

html = urlopen(url)
bs = BeautifulSoup(html, 'html.parser')
images = bs.find_all('img', {'src':re.compile('.jpg')})
for image in images: 
    print(image['src']+'\n')
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
````

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen(url)
bsObj = BeautifulSoup(html)
for link in bsObj.findAll("a"):
  if 'href' in link.attrs:
    print(link.attrs['href'])
````

````python
handle = input("Input a twitter handle")
url = f"https://twitter.com/{handle}?lang=en"
response = requests.get(url).content
soup = BeautifulSoup(response, "lxml")

try :
    what = soup.find("li", {"class":"ProfileNav-item--tweets"})
    tweet = what.find("span", {"class":"ProfileNav-value"})
    total =tweet.get('data-count')
    print(f"Total number of tweets for {handle} is {total}. Wow !")
except :
    print("This twitter handle doesn't exist")
````

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
response = requests.get(url).content
soup = BeautifulSoup(response, "lxml")
````

````python
for i in range(1,11):    
    firstlang = soup.find("div", class_=f"central-featured-lang lang{i}")
    
    langname = firstlang.strong.text
    print(langname)
    articlenb = firstlang.small.bdi.text.replace("\xa0","")   #or split and join
    print(articlenb)
````
